```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option("detach",True)
chrome_options.add_argument('--ignore-certificate-errors')


driver = webdriver.Chrome(options=chrome_options)

driver.get("https://slickdeals.net/")
time.sleep(30)
item_boxes = driver.find_elements(By.CSS_SELECTOR, '.frontpageGrid__feedItem  .dealCard')
list_of_items = []
for item in item_boxes:
    item_list = []
    item_title = item.find_element(By.CSS_SELECTOR, '.dealCard__content .dealCard__title')
    new_item_price = item.find_element(By.CSS_SELECTOR,'.dealCard__content .dealCard__price')
    original_item_price = item.find_element(By.CSS_SELECTOR,'.dealCard__content .dealCard__originalPrice')
    fire_icon = 0
    try:
        item.find_element(By.CSS_SELECTOR, '.dealCard__priceContainer .dealCard__fireIcon')
    except:
        fire_icon = 0
    else:
        fire_icon = 1
    product_img_url = item.find_element(By.CSS_SELECTOR, '.dealCard__imageContainer img')
    time_upload = item.find_element(By.CLASS_NAME, 'dealCard__statusTimestamp')
    store_name = item.find_element(By.CLASS_NAME,'dealCard__storeLink')

    item_list.append(item_title.text) 
    item_list.append(new_item_price.text) 
    item_list.append(original_item_price.text)
    item_list.append(fire_icon)
    item_list.append(product_img_url.get_attribute('src'))
    item_list.append(time_upload.text)
    item_list.append(store_name.text)
    list_of_items.append(item_list)

df_items = pd.DataFrame(list_of_items, columns=['Product','New Price', 'Original Price','Hot Favourite','Product image URL','Upload Time','Store Name'])
```
